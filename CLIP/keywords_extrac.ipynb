```python


# Load the CSV file
csv_path = "emb_sahar/keywords_extractions_with_embeddings1.csv"  # Update with your actual path
df = pd.read_csv(csv_path)

```

```python
keyword_columns = df.columns[2:11]  

all_keywords = df[keyword_columns].values.flatten()

all_keywords = [kw for kw in all_keywords if pd.notna(kw)]

# Count occurrences of each keyword
keyword_counts = Counter(all_keywords)

# Get the top 10 most common keywords
top_keywords = keyword_counts.most_common(50)

# Display results
print("Top Keywords:")
for keyword, count in top_keywords:
    print(f"{keyword}: {count}")

```

```python

keyword_columns = df.columns[2:11]  # Exclude the filename column

# Flatten all keyword values into a single list
all_keywords = df[keyword_columns].values.flatten()

# Remove NaN values if any
all_keywords = [kw for kw in all_keywords if pd.notna(kw)]

# Split multi-word keywords into individual words
all_words = []
for keyword in all_keywords:
    all_words.extend(keyword.split())  # Split by space

# Count occurrences of each word
word_counts = Counter(all_words)

# Get the top 10 most common words
top_words = word_counts.most_common(51)  # Change 10 if needed

# Display results
print("Top Words:")
for word, count in top_words:
    print(f"{word}: {count}")
```

```python
filtered_top_words = [(word, count) for word, count in top_words if word != '-']
filtered_top_words
```

```python
# Plot bar chart
words, counts = zip(*filtered_top_words)

plt.figure(figsize=(10, 5))
bars = plt.bar(words, counts, color='skyblue')

# Add numbers on top of bars
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), str(count), 
             ha='center', va='bottom', fontsize=12, fontweight='bold', rotation=90)

plt.xlabel("Words", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Top 50 Most Frequent Words", fontsize=14)
plt.xticks(rotation=90)  # Rotate labels for readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```

```python
import os
import pandas as pd
from collections import Counter

# Path to your folder containing CSV files
csv_folder = "emb_sahar"

# Counter to collect all words
all_words_counter = Counter()

# Loop through all CSV files in the folder
for filename in os.listdir(csv_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(csv_folder, filename)
        print(file_path)
        df = pd.read_csv(file_path)

        # Select keyword columns (columns 2 to 11)
        keyword_columns = df.columns[2:11]

        # Flatten all keyword values into a list
        all_keywords = df[keyword_columns].values.flatten()

        # Remove NaN values
        all_keywords = [kw for kw in all_keywords if pd.notna(kw)]
        
#         all_words_counter.update(all_keywords)
        # Split multi-word keywords into individual words and update counter
        for keyword in all_keywords:
            words = keyword.split()  # Split by space
            all_words_counter.update(words)

# Get the top 30 most common words
top_words = all_words_counter.most_common(30)

# Display results
print("Top 30 Words:")
for word, count in top_words:
    print(f"{word}: {count}")
```

```python
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(all_words_counter)

# Plot word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Emb_sahar", fontsize=14)

# Show plot
plt.show()
```
